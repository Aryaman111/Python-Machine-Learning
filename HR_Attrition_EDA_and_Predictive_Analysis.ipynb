```python
#create a new column to store the age values. This is only to put the age columns at the end of the data set.So that the target column/dependent variable "Attrition" is placed first
df['Age_Years'] = df['Age']
#Remove the first column called age
df = df.drop('Age', axis = 1)
#Show the dataframe
df
```

```python
#Transform non-numeric columns into numerical columns to easily use use mathematical calculations to learn patterns and make predictions.

from sklearn.preprocessing import LabelEncoder

for column in df.columns:
        if df[column].dtype == np.number:
            continue
        # Fit the LabelEncoder to the column and transform the values
        df[column] = LabelEncoder().fit_transform(df[column])
```

```python
#Split the data into independent 'X' and dependent 'Y' variables
X = df.iloc[:, 1:df.shape[1]].values
Y = df.iloc[:, 0].values #for attribute column
```

```python
# Split the dataset into 75% Training set and 25% Testing set
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.25, random_state = 0)
```

```python
#Use Random Forest Classification algorithm due to its High accuracy and robustness to overfitting:
from sklearn.ensemble import RandomForestClassifier
forest = RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 0)
forest.fit(X_train, Y_train)
```

```python
#Get the accuracy on the training data
forest.score(X_train, Y_train)
```

```python
#This means model is about 98% accurate on the training data
```

```python
#Show the confusion matrix and accuracy for  the model on the test data
#Classification accuracy is the ratio of correct predictions to total predictions made.

from sklearn.metrics import confusion_matrix

cm = confusion_matrix(Y_test, forest.predict(X_test))

TN = cm[0][0]  #True Negatives
TP = cm[1][1]  #True Positives
FN = cm[1][0]  #False Negatives
FP = cm[0][1]  #False Positives

print(cm)
print('Model Testing Accuracy = "{}!"'.format(  (TP + TN) / (TP + TN + FN + FP)))
```

```python
#The model correctly identified 85.59% of the employees that left the company.
```

```python
# Return the feature/column importances in the dataset given (the higher value, the more important the feature or column).

importances = pd.DataFrame({'feature':df.iloc[:, 1:df.shape[1]].columns,'importance':np.round(forest.feature_importances_,3)}) #Note: The target column is at position 0
importances = importances.sort_values('importance',ascending=False).set_index('feature')
importances
```
